```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas 
import numpy as np
```

```python
number_of_rows = 100000
data = pandas.read_csv('data10',sep = '|',header = None, nrows = number_of_rows)
data.head(10)
```

Create a hash table for saving information about represented articles. Hash table looks like: key is tuple (user, article) and values is array[number of showing, number of clicks]. After for each key I compute CTR!

Implementation with hash tables

```python
regret_info = {}
for line in xrange(number_of_rows):
    user = data[1][line]
    article = data[0][line].split(' ')[1]
    click = int(data[0][line].split(' ')[2])
    user_article = (user, article)
    if user_article in regret_info:
        show_info = regret_info[user_article]
        show_info[0] = show_info[0] + 1
        show_info[1] = show_info[1] + click
        regret_info[user_article] = show_info
    else:
        regret_info[user_article] = [1, click]
```
